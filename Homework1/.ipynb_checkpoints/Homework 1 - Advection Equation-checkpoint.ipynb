# 1.0 ADVECTION EQUATION

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Here, you can choose which kind of implementation you want:
- In the first cell, the value of parameters such as a, the courant factor, the observation time, the length of the space coordinates and the number of space points
- In the second cell, from the previous parameters you can create the initial function (selecting the gaussian or the step function by changing 'case'.

To check that the case selected is the right one, a plot of the initial function is shown.

Possible changes: nx, cf, case (gaussian or step)

```python
#DEFINING GENERAL PROPERTIES OF NUMERICAL CODE & ADVECTION EQUATION
a = 1.0
cf = 1.
t_final = 20.0
L = 10.0     
nx = 101 
points_folder = str(nx) + ' points/'
```

```python
#DEFINING INITIAL FUNCTION u(0, x)
case = 'step'  
x = np.linspace(0,10, nx)
dx = x[1]-x[0]
dt = cf*dx/a
t = np.arange(0,20, dt)
if case == 'gaussian':
    x0 = 5 
    u_initial = np.exp(-(x-x0)**2)
else:
    u_initial = []
    for ex in x:
        if (ex>4) and (ex<6):
            u_initial.append(1)
        else:
            u_initial.append(0)
u_initial = np.array(u_initial)
plt.plot(x, u_initial)
plt.show()
```

### 1.0.1: FTCS

```python
u_current = u_initial.copy()
L2_norm = []

if case == 'gaussian':
    case_folder = 'imagesFTCS/'
else:
    case_folder = 'imagesFTCS_stepfunc/'
for idx, time in enumerate(t):
    #reminder: np.roll(u_current, -1) = u(n,j+1) and np.roll(u_current, +1) = u(n,j-1)
    u_next = u_current - a*dt/(2*dx)*(np.roll(u_current, -1) - np.roll(u_current, 1))
    u_current = u_next.copy()
    L2_norm.append(np.sqrt(sum(u_current**2)/len(u_current)))

    """if idx%10 == 0:  
        ax = plt.gca()
        ax.plot(x, u_current)
        ax.set_title('Time='+str(round(time,2)))
        plt.savefig(case_folder+points_folder+'img{:0>5}.png'.format(int(idx/10)), dpi = 300)
        #plt.show()
        ax.clear()"""

plt.plot(t, L2_norm, label = 'my data')
plt.title('L2 Norm with FTCS')
plt.xlabel('time')
plt.ylabel(r'$L_2$ Norm')
check = pd.read_csv('l2norm_FTCS.dat', sep ='\t', header = None)
plt.scatter(check[0], check[1], label = 'check data', color = 'black', s=10)
plt.legend()
plt.savefig(case_folder+points_folder+'L2 norm.png', dpi = 300)
plt.show()
```

```python
plt.plot(x, u_initial/max(u_initial), label = 'Initial Function')
plt.plot(x, u_current/max(u_current), label = 'Evolved Function')
plt.xlabel('x')
plt.ylabel(r'u/max(u)')
plt.legend()
plt.title('Comparison between initial function and its evolution')
plt.savefig(case_folder+points_folder+'comparison for FTCS.png', dpi = 300)
```

### 1.0.2: Lax-Friedrichs

```python
u_current = u_initial.copy()
L2_norm = []

if case == 'gaussian':
    case_folder = 'imagesLAX-FRIEDRICHS/'
else:
    case_folder = 'imagesLAX-FRIEDRICHS_stepfunc/'
    
for idx, time in enumerate(t):
    #reminder: np.roll(u_current, -1) = u(n,j+1) and np.roll(u_current, +1) = u(n,j-1)
    u_next = 1/2*(np.roll(u_current, -1) + np.roll(u_current, 1)) - a*dt/(2*dx)*(np.roll(u_current, -1) - np.roll(u_current, 1))
    u_current = u_next.copy()
    L2_norm.append(np.sqrt(sum(u_current**2)/len(u_current)))

    """if idx%10 == 0:  
        ax = plt.gca()
        ax.plot(x, u_current)
        ax.set_title('Time='+str(round(time,2)))
        plt.savefig(case_folder+points_folder+'img{:0>5}.png'.format(int(idx/10)), dpi = 300)
        #plt.show()
        ax.clear()"""

plt.plot(t, L2_norm, label = 'my data')
plt.title('L2 Norm with Lax-Friedrichs')
plt.xlabel('time')
plt.ylabel(r'$L_2$ Norm')
check = pd.read_csv('l2norm_LAX.dat', sep ='\t', header = None)
plt.scatter(check[0], check[1], label = 'check data', color = 'black', s=10)
plt.legend()
plt.savefig(case_folder+points_folder+'L2 norm_cf_15.png', dpi = 300)
plt.show()

plt.plot(x, u_initial/max(u_initial), label = 'Initial Function')
plt.plot(x, u_current/max(u_current), label = 'Evolved Function')
plt.legend()
plt.xlabel('x')
plt.ylabel('u/max(u)')
plt.title('Comparison between initial function and its evolution')
plt.savefig(case_folder+points_folder+'Comparison LAXFRIED_cf_15.png', dpi = 300)
```

### 1.0.3: Leapfrog

```python
u_current = u_initial.copy()
L2_norm = []

u_next = 1/2*(np.roll(u_current, -1) + np.roll(u_current, 1)) - a*dt/(2*dx)*(np.roll(u_current, -1) - np.roll(u_current, 1))
u_previous = u_current.copy()
u_current = u_next.copy()
L2_norm.append(np.sqrt(sum(u_current**2)/len(u_current)))


if case == 'gaussian':
    case_folder = 'imagesLEAPFROG/'
else:
    case_folder = 'imagesLEAPFROG_stepfunc/'
    
for idx, time in enumerate(t):
    if time != 0:
        #reminder: np.roll(u_current, -1) = u(n,j+1) and np.roll(u_current, +1) = u(n,j-1)
        u_next = u_previous - a*dt/(2*dx)*(np.roll(u_current, -1) - np.roll(u_current, 1))
        u_previous = u_current.copy()
        u_current = u_next.copy()
        L2_norm.append(np.sqrt(sum(u_current**2)/len(u_current)))
    
        """if idx%10 == 0:  
            ax = plt.gca()
            ax.plot(x, u_current)
            ax.set_title('Time='+str(round(time,2)))
            plt.savefig(case_folder+points_folder+'img{:0>5}.png'.format(int(idx/10)), dpi = 300)
            #plt.show()
            ax.clear()"""

plt.plot(t, L2_norm/L2_norm[0], label = 'my data', alpha = 0.3)
plt.title('L2 Norm with Leapfrog')
plt.xlabel('time')
plt.ylabel(r'$L_2$ Norm')
plt.axhline(1, color = 'red', alpha = 0.7, linestyle = '--')
check = pd.read_csv('l2norm_LEAPFROG.dat', sep ='\t', header = None)
plt.scatter(check[0], check[1]/check[1][0], label = 'check data', color = 'black', s=10)
if nx == 1001:
    plt.legend(title = 'Note: +1 on the upper left')
else:
    plt.legend()
plt.savefig(case_folder+points_folder+'L2 norm_leapfrog.png', dpi = 300)
plt.show()

plt.plot(x, u_initial, label = 'Initial Function')
plt.plot(x, u_current, label = 'Evolved Function')
plt.legend()
plt.xlabel('x')
plt.ylabel('u')
plt.title('Comparison between initial function and its evolution')
plt.savefig(case_folder+points_folder+'Comparison leapfrog.png', dpi = 300)
```

### 1.0.4: Lax-Wendroff

```python
u_current = u_initial.copy()
L2_norm = []


if case == 'gaussian':
    case_folder = 'imagesLAXWENDROFF/'
else:
    case_folder = 'imagesLAXWENDROFF_stepfunc/'

for idx, time in enumerate(t):
    u_next = u_current - a*dt/(2*dx)*(np.roll(u_current, -1) - np.roll(u_current, 1))+ a**2*dt**2/(2*dx**2)*(np.roll(u_current, -1)-2*u_current + np.roll(u_current, +1))
    u_current = u_next.copy()
    L2_norm.append(np.sqrt(sum(u_current**2)/len(u_current)))

    """if idx%10 == 0:  
        ax = plt.gca()
        ax.plot(x, u_current)
        ax.set_title('Time='+str(round(time,2)))
        plt.savefig(case_folder+points_folder+'img{:0>5}.png'.format(int(idx/10)), dpi = 300)
        #plt.show()
        ax.clear()"""

plt.plot(t, L2_norm/L2_norm[0], label = 'my data')
plt.title('L2 Norm with Lax-Wendroff')
plt.xlabel('time')
plt.ylabel(r'$L_2$ Norm')
check = pd.read_csv('l2norm_LAXWENDROFF.dat', sep ='\t', header = None)
plt.scatter(check[0], check[1]/check[1][0], label = 'check data', color = 'black', s=10)
plt.legend()
plt.savefig(case_folder+points_folder+'L2 norm_LAXWENDR_cf_1.png', dpi = 300)
plt.show()

plt.plot(x, u_initial, label = 'Initial Function')
plt.plot(x, u_current, label = 'Evolved Function')
plt.legend()
plt.xlabel('x')
plt.ylabel('u')
plt.title('Comparison between initial function and its evolution')
plt.savefig(case_folder+points_folder+'Comparison LAXWENDROFF_cf_1.png', dpi = 300)
```
