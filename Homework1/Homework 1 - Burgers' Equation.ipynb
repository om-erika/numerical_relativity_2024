# 1.3 BURGERS' EQUATION

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Here, it is possible to choose which kind of implementation you want:
- The value of parameters such as a, the courant factor, the observation time, the length of the space coordinates and the number of space points
- The way the ODE is written (selecting the flux conservative or non flux conservative scenario by changing 'case').

To check that the case selected is the right one, a plot of the initial function is shown.

Possible changes: nx, cf, scenario (conservative or non conservative)

```python
#DEFINING GENERAL PROPERTIES OF NUMERICAL CODE & ADVECTION EQUATION
cf = 0.5
t_final = 0.5
L = 10.0     
nx = 101 
total = True
case = 'non conservative'      #options: "conservative" - "non conservative"
case_folder = 'burger/' + case + '/'
points_folder = str(nx) + ' points/'
```

```python
#DEFINING INITIAL FUNCTION u(0, x)
x = np.linspace(0,10, nx)
dx = x[1]-x[0]
#dt = cf*dx/a
time = [] #np.arange(0,t_final, dt)
x0 = 5 
u_initial = 10*np.exp(-(x-x0)**2)
u_initial = np.array(u_initial)
a = max(u_initial)
plt.plot(x, u_initial)
plt.show()
```

### 1.0.1: CONSERVATIVE or NON CONSERVATIVE version

```python
u_current = u_initial.copy()
L2_norm = []
time = []
if case == 'conservative':
    case_folder = 'burger/cons_flux/'
else:
    case_folder = 'burger/non_cons_flux/'

t = 0
time.append(t)
L2_norm.append(np.sqrt(sum(u_current**2)/len(u_current)))
u_total = []
u_total.append(u_current.copy())
check = 0
while t < t_final:
    dt = cf*dx/np.max(u_current)  #change dt as cf*dx/max(u(x,t))
    #reminder: np.roll(u_current, -1) = u(n,j+1) and np.roll(u_current, +1) = u(n,j-1)
    if case == 'conservative':
        u_next = u_current - (dt/dx)*(0.5*(u_current**2) - 0.5*((np.roll(u_current, 1))**2))     
    else: 
        u_next = u_current - dt/dx*u_current*(u_current-np.roll(u_current, +1))
    u_current = u_next.copy()
    if (t > t_final/2) and check == 0:
        u_med = u_current.copy()
        check = 1
    t = t+dt
    if total:
        u_total.append(u_current.copy())
    time.append(t)
    L2_norm.append(np.sqrt(sum(u_current**2)/len(u_current)))
```

```python
if case == 'conservative':
    u_current_cons = u_current.copy()
    u_med_cons = u_med.copy()
    u_total_cons = u_total.copy()
    time_cons = time.copy()
if case == 'non conservative':
    u_current_non_cons = u_current.copy()
    u_med_non_cons = u_med.copy()
    u_total_non_cons = u_total.copy()
    time_non_cons = time.copy()

plt.plot(time, np.array(L2_norm)/np.array(L2_norm[0]), label = r'$L_2$ Norm')
plt.title('L2 Norm for ' + str(case) + ' flux version')
plt.xlabel('time')
plt.ylabel(r'$L_2$ Norm')
plt.legend()
plt.savefig(case_folder+points_folder+'L2 norm.png', dpi = 300)
plt.show()

plt.plot(x, u_initial, label = 'Initial Function')
plt.plot(x, u_current, label = 'Evolved Function')
plt.xlabel('x')
plt.ylabel(r'u/max(u)')
plt.legend()
plt.title('Comparison between initial function and its evolution')
plt.savefig(case_folder+points_folder+'comparison for '+ case +'.png', dpi = 300)
```
