# SOD SHOCK TUBE

```python
import matplotlib.pyplot as plt
import numpy as np
from kuibit.simdir import SimDir
from kuibit.grid_data import UniformGrid
```

```python
#initial data from the Exact Riemann solver of https://www.brunogiacomazzo.org/?page_id=395
exact = np.loadtxt("Sod_exact.dat")
```

```python
x_exact=exact[:,0]
rho_exact=exact[:,1]
press_exact=exact[:,2] #note: this is pgas+pmag
vx_exact=exact[:,3]
vy_exact=exact[:,4]
vz_exact=exact[:,5]
By_exact=exact[:,6]
Bz_exact=exact[:,7]
```

```python
plt.plot(x_exact, rho_exact, label="exact")
plt.xlim(-0.5, 0.5)
plt.legend()
plt.show()
```

## Read the Numerical Results

```python
gf_100= SimDir("Sod_1d_100/").gf
gf_400= SimDir("./Sod_1d_400/").gf
gf_1000= SimDir("./Sod_1d_1000/").gf
```

```python
vars1d_100=gf_100.x
vars1d_400=gf_400.x
vars1d_1000=gf_1000.x
```

```python
rho_100=vars1d_100.fields.rho
rho_400=vars1d_400.fields.rho
rho_1000=vars1d_1000.fields.rho

press_100=vars1d_100.fields.press
press_400=vars1d_400.fields.press
press_1000=vars1d_1000.fields.press

dens_100=vars1d_100.fields.vel[0]
dens_400=vars1d_400.fields.vel[0]
dens_1000=vars1d_1000.fields.vel[0]
```

```python
small_grid_100 = UniformGrid([101], x0=[-0.45], x1=[0.45])
small_grid_400 = UniformGrid([101], x0=[-0.45], x1=[0.45])
small_grid_1000 = UniformGrid([101], x0=[-0.45], x1=[0.45])
```

```python
#rho_x.iterations[-1]
rho_final_100 = rho_100.read_on_grid(100, small_grid_100)
rho_final_400 = rho_400.read_on_grid(200, small_grid_400)
rho_final_1000 = rho_1000.read_on_grid(500, small_grid_1000)
```

```python
plt.plot(*rho_final_100.coordinates_meshgrid(), rho_final_100.data_xyz, label="100 bins")
plt.plot(*rho_final_400.coordinates_meshgrid(), rho_final_400.data_xyz, label="400 bins")
plt.plot(*rho_final_1000.coordinates_meshgrid(), rho_final_1000.data_xyz, label="1000 bins")

plt.plot(x_exact, rho_exact, label="exact", alpha = 0.8)
plt.xlim(-0.5, 0.5)
#plt.ylim(0.1, 0.3)
plt.xlabel('x')
plt.ylabel(r'$\rho$')
plt.title('Density')
plt.legend()
plt.savefig('Comparison density SOV.png', dpi = 300)
plt.show()
```

```python
dens_final_100 = dens_100.read_on_grid(100, small_grid_100)
dens_final_400 = dens_400.read_on_grid(200, small_grid_400)
dens_final_1000 = dens_1000.read_on_grid(500, small_grid_1000)
```

```python
plt.plot(*dens_final_100.coordinates_meshgrid(), dens_final_100.data_xyz, label="100 bins")
plt.plot(*dens_final_400.coordinates_meshgrid(), dens_final_400.data_xyz, label="400 bins")
plt.plot(*dens_final_1000.coordinates_meshgrid(), dens_final_1000.data_xyz, label="1000 bins")

plt.plot(x_exact, vx_exact, label="exact", alpha = 0.8)
plt.xlim(-0.5, 0.5)
#plt.ylim(0.1, 0.3)
plt.xlabel('x')
plt.ylabel(r'$v_x$')
plt.title('Velocity Along x-Axis')
plt.legend()
plt.savefig('Comparison velocity SOV.png', dpi = 300)
plt.show()
```

```python
press_final_100 = press_100.read_on_grid(100, small_grid_100)
press_final_400 = press_400.read_on_grid(200, small_grid_400)
press_final_1000 = press_1000.read_on_grid(500, small_grid_1000)

plt.plot(*press_final_100.coordinates_meshgrid(), press_final_100.data_xyz, label="100 bins")
plt.plot(*press_final_400.coordinates_meshgrid(), press_final_400.data_xyz, label="400 bins")
plt.plot(*press_final_1000.coordinates_meshgrid(), press_final_1000.data_xyz, label="1000 bins")

plt.plot(x_exact, press_exact, label="exact", alpha = 0.8)
plt.xlim(-0.5, 0.5)
#plt.ylim(0.1, 0.3)
plt.xlabel('x')
plt.ylabel('P')
plt.title('Pressure')
plt.legend()
plt.savefig('Comparison pressure SOV.png', dpi = 300)
plt.show()
```

```python
press_initial_1000 = press_1000.read_on_grid(0, small_grid_100)
dens_initial_1000 = dens_1000.read_on_grid(0, small_grid_400)
rho_initial_1000 = rho_1000.read_on_grid(0, small_grid_1000)

plt.plot(*press_initial_1000.coordinates_meshgrid(), press_initial_1000.data_xyz)
plt.xlabel('x')
plt.ylabel('P')
plt.title('Initial Pressure')
plt.savefig('Initial pressure SOV.png', dpi = 300)
plt.show()

plt.plot(*dens_initial_1000.coordinates_meshgrid(), dens_initial_1000.data_xyz)
plt.xlabel('x')
plt.ylabel(r'$v_x$')
plt.title('Initial Velocity Along x-Axis')
plt.savefig('Initial velocity along x axis SOV.png', dpi = 300)
plt.show()

plt.plot(*rho_initial_1000.coordinates_meshgrid(), rho_initial_1000.data_xyz)
plt.xlabel('x')
plt.ylabel(r'$\rho$')
plt.title('Initial Density')
plt.savefig('Initial density SOV.png', dpi = 300)
plt.show()
```
